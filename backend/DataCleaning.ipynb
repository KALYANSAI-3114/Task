2. Data Cleaning

My approach

1. Finding Duplicate values if there are duplicate values then remove them
2. Finding missing values, if there are missing values then they can be removed or filled 

```python
#Duplicate values
shipping_data.duplicated().sum()
```

```python
shipping_data[shipping_data.duplicated()] # This will show what values is repeated 
```

there is a duplicate value in the excel as it is only one duplicate value we can remove it from excel file 

```python
shipping_data.drop_duplicates(inplace=True) # It will drop the duplicate values 
#Now we can check

shipping_data.duplicated().sum()
```

```python
shipping_data.isnull().sum()
```

There are so many missing values we want to handle all the missing values

```python
for col in shipping_data.columns:
    missing_percentage = shipping_data[col].isnull().mean() * 100
    print(f"{col}: {missing_percentage:.2f}% missing")
```

We can remove the features  which are more than 50% 

```python
# I am taking threshold value as 0.5 and less than that remove them

threshold_value = 0.5 #50%
shipping_data = shipping_data.loc[:, shipping_data.isnull().mean() < threshold_value]
shipping_data.isnull().sum()
```

```python
for col in shipping_data.columns:
    missing_percentage = shipping_data[col].isnull().mean() * 100
    print(f"{col}: {missing_percentage:.2f}% missing")
```

Still we have missing values so as per the requirement we can keep "Null","Unknown" values to the missing ones

```python
shipping_data.head()
```

```python
shipping_data[shipping_data['Ref #'].isnull()]
```

As per the task requirements there are some features that are to be unchanged they are 
1. Ref #
2. Tracking #	
3. Cost Columns 
4. To ID	

These 4 are to nan only as we dont required them in the task

Fill the missing values in the categorical columns as "Unknown"

```python
categorical_columns = [
    "Carrier",
    "Service Level",
    "Shipment Type",
    "Payment Type",
    "Location",
    "To Company"
]
shipping_data[categorical_columns] = shipping_data[categorical_columns].fillna("Unknown")
```

```python
shipping_data['Delivery Status'].unique()
# As we have all the values like shipped,canceled and all if it is pending then only it will be nan
#So convert nan - pending
```

```python
shipping_data['Delivery Status'] = shipping_data['Delivery Status'].fillna("pending")
```

```python
shipping_data['Last Updated By'].unique()
```

As there are 15 person names we cannot blindly choose a name from these 15 members and fill the missing values so i will keep in place of the missing value "Unknown user"

```python
shipping_data['Last Updated By'] = shipping_data["Last Updated By"].fillna("Unknown User")
```

```python
shipping_data['Last Updated Date'].isnull().sum()
```

There are missing values in last updated date feature so to handle this we can take ship Date as Last Updated Date because 
It means no updates happened after creation,so the creation time (Ship Date) is the correct fallback

```python
shipping_data['Last Updated Date'] = shipping_data['Last Updated Date'].fillna(shipping_data['Ship Date'])
```

```python
shipping_data.isnull().sum()
```

```python
cols = ['Ref #','Tracking #','To ID']
for col in cols:
    shipping_data[col] = shipping_data[col].fillna("Unknown")
```

```python
shipping_data.isnull().sum()
```
